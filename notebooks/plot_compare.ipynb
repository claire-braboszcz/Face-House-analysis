All the subjects compared.

<div class="alert alert-danger"><h4>Warning</h4><p>This example will open a lot of figures simultaneously!</p></div>



```python
import os.path as op
import sys
import os

import numpy as np

import matplotlib.pyplot as plt

import mne

from autoreject import set_matplotlib_defaults


set_matplotlib_defaults(plt)

list_evoked = list()
exclude = [7]  # Excluded subjects
event_id = {'stim/face':101, 'stim/house':102, 'imag/face':201, 'imag/house':202 }


ana_path = '/home/claire/DATA/Data_Face_House_new_proc/'
data_folder= '/EEG/Preproc'

for subject_id in range(1, 25):
    if subject_id in exclude:
        continue
    subject = 'S%02d' %subject_id
    data_path = os.path.join(ana_path + subject + data_folder)
    fname_in = os.path.join(data_path, '%s-clean-epo.fif' %subject)
    epochs=mne.read_epochs(fname_in)
    epochs.interpolate_bads(reset_bads=True)
    evoked = {cond:list(epochs[cond].iter_evoked()) for cond in event_id}

    list_evoked.append(evoked)
    
times = np.arange(0.1, 0.26, 0.025)

plt.rcParams['figure.max_open_warning'] = 200
```

Evoked responses on EEG and MEG, see
`sphx_glr_auto_scripts_07-make_evoked.py`.



```python
for idx, evoked in enumerate(list_evoked):
        for cond in event_id:
            evoked[cond].plot_joint(title='Subject %s %s' % (idx + 1, cond),
                                    show=False)
plt.show()
```
